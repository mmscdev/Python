```
nota1 = float(input("Digite a nota bimestral 1:"))
nota2 = float(input("Digite a nota bimestral 2:"))
nota3 = float(input("Digite a bimestral nota 3:"))
nota4 = float(input("Digite a bimestral nota 4:"))
lista_notas = [nota1,nota2,nota3,nota4]
media =  sum(lista_notas)/len(lista_notas)
mediaOutraForma =  (nota1+nota2+nota3+nota4)/4
print(f"A média bimestral é: Com lista {media} Modo simples:{mediaOutraForma}")
```

Crie um algoritmo em Python que peça o raio de um círculo, calcule e mostre sua área.

```
#Área do círculo arrendondada com precisão de duas casas decimais
def CalcularAreaCirculo(raio):
   return round(3.1415926535898*(raio)**2,2)

raio = float(input("Digite o raio do círculo em metros: "))
area = CalcularAreaCirculo(raio) 
print(f"A área do círculo de raio {raio} é : {area}m²")
```

Crie um algoritmo em Python que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário.

```
#Área do quadrado arrendondada com precisão de duas casas decimais
def CalcularAreaQuadrado(lado):
   return round((lado)**2,2)

lado = float(input("Digite o lado do quadrado em metros: "))
area = CalcularAreaQuadrado(lado) 
areaDobro = area * 2
print(f"A área do quadrado de lado {lado} é : {area}m² e o dobro da área é: {areaDobro} m²")
```

Crie um algoritmo em Python que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês.


```
def CalcularSalarioMensal(homemHora,totalHorasMes):
   return homemHora* totalHorasMes

homemHora = float(input("Digite quanto você ganha por hora: "))
totalHorasMes = float(input("Digite o número de horas trabalhadas no mês: "))
salarioMensal = CalcularSalarioMensal(homemHora,totalHorasMes) 
print(f"O salário no referido mês é : R$ {salarioMensal}")
```

Crie um algoritmo em Python que peça a temperatura em graus Farenheit, transforme e mostre a temperatura em graus Celsius.
C = (5 * (F-32) / 9).

```
def ConverterFarenheitCelsius(temperaturaF):
   return (5*(temperaturaF - 32))

temperaturaF = float(input("Digite a temperatura em graus Farenheit: "))
temperaturaCelsius = ConverterFarenheitCelsius(temperaturaF)
print(f"A temperatura {temperaturaF} F corresponde a {temperaturaCelsius} C ")
```

Crie um algoritmo em Python que peça a temperatura em graus Celsius, transforme e mostre em graus Farenheit.

```
def ConverterCelsiusFarenheit(temperaturaC):
   return (temperaturaC *(9/5)) + 32

temperaturaC = float(input("Digite a temperatura em graus Celsius: "))
temperaturaF = ConverterCelsiusFarenheit(temperaturaC)
print(f"A temperatura {temperaturaCelsius} C corresponde a {temperaturaF} F")
```
