Crie um algoritmo em Python que peça 2 números e imprima o maior deles.

```
def ObterMaior(num1,num2):
  if(num1 > num2):
    return num1
  else:
    return num2

num1 = float(input("Digite um número: "))
num2 = float(input("Digite outro número: "))
maior = ObterMaior(num1,num2)
print(f"O maior número digitado é : {maior}")
```

Crie um algoritmo em Python que verifique se uma letra digitada é vogal ou consoante.

```
def ValidarLetra(letra):
  vogais =  ['a','e','i','o','u']
  if(letra.lower() in vogais): return "é vogal."
  else: return "é consoante."
  
letra = str(input("Digite uma letra: "))
print(f"A letra digitada: {letra} {ValidarLetra(letra)}")
```


Faça um programa para a leitura de duas notas parciais de um aluno. O programa deve calcular a média alcançada por aluno e apresentar:
* "Aprovado", se a média alcançada for maior ou igual a sete;
* "Reprovado", se a média for menor do que sete;
* "Aprovado com Distinção", se a média for igual a 10.

```
def CalcularMedia(nota1,nota2):
   media = (nota1+nota2)/2
   if(media >= 7): return f"A média é {media}. Você está Aprovado"
   if(media <= 7): return f"A média é {media}. Você está Reprovado"
   if(media == 10): return f"A média é {media}. Você está Aprovado com Distinção"

nota1 = float(input("Digite a nota 1: "))
nota2 = float(input("Digite a nota 2: "))
print(CalcularMedia(nota1,nota2))
```

Crie um algoritmo em Python que leia três números e mostre o maior deles.

```
def ObterMaior(num1,num2,num3):
  if(num1 > num2 and num1 > num3):
    return num1
  elif(num2 > num1 and num2 > num3):
    return num2
  elif(num3 > num1 and num3 > num2):
    return num3

num1 = float(input("Digite um número: "))
num2 = float(input("Digite outro número: "))
num3 = float(input("Digite outro número: "))
maior = ObterMaior(num1,num2,num3)
print(f"O maior número digitado é : {maior}")
```

Faça um programa que pergunte o preço de 5 produtos e informe qual produto você deve comprar, sabendo que a decisão é sempre pelo mais barato.

```
listaProdutos = ["Produto 1","Produto 2","Produto 3","Produto 4","Produto 5"]
lista = list()
for i in range(0,5):
  lista.append(float(input(f"Digite o valor do produto {listaProdutos[i]}: ")))

menorValor = min(lista)
index = lista.index(menorValor)
print(f"O produto mais barato é o {listaProdutos[index]} custa: R$ {menorValor}")
```

As Organizações Mendéz resolveram dar um aumento de salário aos seus colaboradores e lhe contraram para desenvolver o programa que calculará os reajustes em Python. Faça um programa que receba o salário de um colaborador e o reajuste segundo o seguinte critério, baseado no salário atual:

* salários entre 680,00 e 800,00 : aumento de 15%<br>
* salários entre 800,00 e 2500,00 : aumento de 10%<br>
* salários de 2500,00 em diante : aumento de 5% após o aumento ser realizado.

Informe na tela:
* o salário antes do reajuste;
* o percentual de aumento aplicado;
* o valor do aumento;
* o novo salário, após o aumento.

```
def CalcularReajuste(salario):
  if(salario >= 680 and salario <= 800): return f"Seu salário reajustado com 15% é : {salario + (salario * (15/100))} em relação ao salário atual {salario}"
  if(salario > 800 and salario <= 2500): return f"Seu salário reajustado com 10% é : {salario + (salario * (10/100))} em relação ao salário atual {salario}"
  if(salario > 2500): return f"Seu salário reajustado com 5% é : {salario + (salario * (5/100))} em relação ao salário atual {salario}"

salarioAtual = float(input("Digite seu salário atual: "))
print(CalcularReajuste(salarioAtual))
```
